```python
asteroids_data=[]
target=10000
url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date=2024-01-01&end_date=2024-01-08&api_key={API_KEY}"

while len(asteroids_data) < target:
    response = requests.get(url)
    data = response.json()
    details = data['near_earth_objects']
    for date,info in details.items():
        for i in info:
            asteroids_data.append(dict(id=i['id'],
                               neo_reference_id=i['neo_reference_id'],
                               name=i['name'],
                               absolute_magnitude_h=i['absolute_magnitude_h'],
                               estimated_diameter_min_km=i['estimated_diameter']['kilometers']['estimated_diameter_min'],
                               estimated_diameter_max_km=i['estimated_diameter']['kilometers']['estimated_diameter_max'],
                               is_potentially_hazardous_asteroid=i['is_potentially_hazardous_asteroid'],
                               close_approach_date=i['close_approach_data'][0]['close_approach_date'],
                              relative_velocity_kmph=i['close_approach_data'][0]['relative_velocity']['kilometers_per_hour'],
                                      astronomical=i['close_approach_data'][0]['miss_distance']['astronomical'],
                                      lunar_distance=i['close_approach_data'][0]['miss_distance']['lunar'],
                                      miss_distance=i['close_approach_data'][0]['miss_distance']['kilometers'],
                                      orbiting_body=i['close_approach_data'][0]['orbiting_body']))
            if len(asteroids_data) >= target:
                break
        if len(asteroids_data) >= target:
            break

url = data['links']['next']
print(url)
        
```

#ASTEROIDS

```python
cursor.execute("""create table asteroids_details(
id INT,
name VARCHAR(100),
absolute_magnitude_h FLOAT,
estimated_diameter_min_km FLOAT,
estimated_diameter_max_km FLOAT,
is_potentially_hazardous_asteroids BOOL
)
""")
```

```python
insert_query="insert into asteroids_details values(%s,%s,%s,%s,%s,%s)"

for i in asteroids_data:
    id = int(i['id'])
    name=i['name']
    mangnitude=float(i['absolute_magnitude_h'])
    dia_min=float(i['estimated_diameter_min_km'])
    dia_max=float(i['estimated_diameter_max_km'])
    hazard=i['is_potentially_hazardous_asteroid']

    values=(id,name,mangnitude,dia_min,dia_max,hazard)
    cursor.execute(insert_query,values)
    connection.commit() 
```

#CLOSE APPROACH TABLE

```python
cursor.execute("""create table close_approach1(
neo_reference_id INT,
close_approach_data DATE,
relative_velocity_kmph FLOAT,
astronomical FLOAT,
miss_distance_km FLOAT,
miss_distance_lunnar FLOAT,
orbiting_body text)
""")
```

```python
from datetime import datetime

insert_query="insert into close_approach1 values(%s,%s,%s,%s,%s,%s,%s)"

for i in asteroids_data:
    id=int(i['id'])
    formatted_date=datetime.strptime(i['close_approach_date'], "%Y-%m-%d").date()
    velocity=float(i['relative_velocity_kmph'])
    astro=float(i['astronomical'])
    miss_dist=float(i['miss_distance'])
    lunar=float(i['lunar_distance'])
    body=i['orbiting_body']
    
    values=(id,formatted_date,velocity,astro,miss_dist,lunar,body)
    cursor.execute(insert_query,values)

connection.commit()
```

#Average velocity of each asteroid over multiple approches

```python
q2="""
select neo_reference_id, avg(relative_velocity_kmph) as avg_velocity_kmph
from close_approach1
where orbiting_body='Earth'
group by neo_reference_id
order by avg_velocity_kmph desc;
"""

cursor.execute(q2)
results = cursor.fetchall()
```

#FIND POTANTIALLY HAZARDOUS ASTEROIDS THAT HAVE APPROACHED EARTH MORE THAN 3 TIMES

```python
q4="""select a.id, a.name, count(c.neo_reference_id) as approached_earth
from asteroids_details a
join close_approach1 c
on a.id=c.neo_reference_id
where a.is_potentially_hazardous_asteroids=1
and c.orbiting_body='Earth'
group by a.id, a.name
having count(c.neo_reference_id)>3
order by approached_earth desc;"""

cursor.execute(q4)
results = cursor.fetchall()
```

```python
#ASTEROIDS WHOSE CLOSEST APPROACH IS GETTING NEARER OVER TIME 
q8="""SELECT 
    ca.neo_reference_id,
    ad.name,
    ca.close_approach_data,
    ca.miss_distance_km
FROM 
    close_approach1 ca
JOIN 
    asteroids_details ad 
    ON ca.neo_reference_id = ad.id
ORDER BY 
    ca.close_approach_data ASC, 
    ca.miss_distance_km ASC;
"""

cursor.execute(q8)
results = cursor.fetchall()
```

Display the name of each asteroid along with the date and miss distance of its closest approach to Earth.

```python
q9="""SELECT 
    ad.name,
    ca.close_approach_data,
    ca.miss_distance_km
FROM 
    close_approach1 ca
JOIN 
    asteroids_details ad 
ON 
    ca.neo_reference_id = ad.id
WHERE 
    (ca.neo_reference_id, ca.miss_distance_km) in (
        SELECT neo_reference_id, MIN(ca2.miss_distance_km)
        FROM close_approach1 ca2
        GROUP BY neo_reference_id
    )
ORDER BY 
    ca.miss_distance_km asc;"""
cursor.execute(q9)
results = cursor.fetchall()

```

Count how many approaches happened per month

```python
q11="""select extract(year from ca.close_approach_data) as year,
        extract(month from ca.close_approach_data) as month,
        count(*) as approach_count
        from close_approach1 ca
        group by year, month
        order by year, month;"""

cursor.execute(q11)
results = cursor.fetchall()
```
